```python
#import pathlib and csv
from pathlib import Path
import csv
from operator import itemgetter

#set the file path
csvpath = Path(r"C:\Users\dhaam\Documents\Fintech Course\Homework\Unit2homework\python-homework\PyBank\budget_data.csv")

#Initialise variable to hold months and profits_losses
months = []
profits_losses = []
total_months = 0
max_profit = 0
max_loss = 0
diff = 0
cumulative_diff = 0
pnlchange = []

#Open the input path as a file object
with open(csvpath, 'r') as csvfile:
    #pass in the csvfile to the csv.reader() function (with ',' as the delimiter/separator) and return the csvreader object 
    csvreader = csv.reader(csvfile, delimiter=',')
   # Read the header row first
    header = next(csvreader)
    previous_row = None
   
    #read each row of data after the header
    for row in csvreader:
        #set Date value equal to the value in the 0 column of each row
        date = str(row[0])
        #set pnl value equal to the value in the 1st column of each row
        pnl = int(row[1])
        #append the row date value to the list of months and the row pnl value to the list of profits_losses
        months.append(date)
        profits_losses.append(pnl)
        
        #this calculates the difference between two rows in the Profits/Losses.
        if previous_row is None:
            diff = 0
        else:
            diff = int(row[1]) - int(previous_row[1])
      
        #add the difference values to a list
        pnlchange.append(diff)
        
        #combine months list and difference values into a single list
        budget_change = list(zip(months, pnlchange))

        #this calculates the cumulative total of the difference
        cumulative_diff = cumulative_diff + diff
        #this keeps track of the row that the difference is calculating 
        previous_row = row
  
        #this calculates the greatest value in the list to determine the greatest increase.
        if diff > max_profit:
            max_profit = diff
        #calculates the lowest value in the list to determine the greatest increase.
        if diff < max_loss:
            max_loss = diff
                           

#calcualte total no. of months in the list.
total_months=len(months)

#create variable for total pnl
total_pnl = sum(profits_losses)

#calculate average change
average = cumulative_diff/(total_months-1) 

#Header
print("FINANCIAL ANALYSIS")
#print total months
print(f"Total Months: {total_months}")
#print total pnl
print(f"Total: {total_pnl}")
#print average change
print(f"Average Change: {average}")
#print greatest increse in profits with corresponding month
print(f"Greatest Increase in Profits: {max(budget_change, key=itemgetter(1))[0]} {max_profit}")
#print greatest decrease in profits with corresponding month
print(f"Greatest Decrease in Profits: {min(budget_change, key=itemgetter(1))[0]} {max_loss}")
```

```python
#set output file name
output_path = 'output.txt'
#openn then output path as a file object
with open(output_path, 'w') as file:
    #write calculations to the output file, convert to string
    file.write("FINANCIAL ANALYSIS")
    file.write(f"Total Months: {total_months}")
    file.write(f"Total: {total_pnl}")
    file.write(f"Average Change: {average}")
    file.write(f"Greatest Increase in Profits: {max(budget_change, key=itemgetter(1))[0]} {max_profit}")
    file.write(f"Greatest Decrease in Profits: {min(budget_change, key=itemgetter(1))[0]} {max_loss}")
```
